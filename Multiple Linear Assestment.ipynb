Q1. Explain the concept of R-squared in linear regression models. How is it calculated, and what does it
represent?

Ans. R-squared (R²), also known as the coefficient of determination, is a statistical metric used to evaluate the goodness of fit of a linear regression model. It represents the proportion of the variance in the dependent variable that is predictable from the independent variables.

What R-squared Represents
Interpretation:

An R² value of 1 indicates that the model explains 100% of the variance in the dependent variable.
An R² value of 0 means the model explains none of the variance, i.e., it is no better than a simple mean.

Q4. What are RMSE, MSE, and MAE in the context of regression analysis? How are these metrics
calculated, and what do they represent?

Ans. Metric	Key Traits	When to Use
MSE	Penalizes larger errors heavily.	When large errors are particularly undesirable.
RMSE	Interpretable in original units.	When you want errors in the same units as the target variable.
MAE	Less sensitive to outliers.	When you want a more robust metric for datasets with outliers.

Q5. Discuss the advantages and disadvantages of using RMSE, MSE, and MAE as evaluation metrics in
regression analysis.

Ans. Metric	Advantages	Disadvantages	Best Use Cases
MSE	Penalizes large errors, easy to optimize.	Hard to interpret, sensitive to outliers.	When large errors are highly undesirable.
RMSE	Interpretable units, emphasizes large errors.	Sensitive to outliers, needs context.	When you want error magnitude in original units.
MAE	Robust to outliers, easy to interpret.	Ignores error variance, harder to optimize.	When errors of all sizes matter equally.
